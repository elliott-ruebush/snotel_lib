# SNOTEL Metadata Extremes

Quick look at highest/lowest stations out there!

```python
import logging

from snotel_lib import SnotelClient
from snotel_lib.utils import get_min_and_max_rows

logging.basicConfig(level=logging.DEBUG, force=True)


client = SnotelClient()
metadata = client.get_stations_metadata(force_update=True)
metadata.head()
```

```python
# Highest and lowest elevation SNOTELs
get_min_and_max_rows(metadata, "elevation_m")
```

```python
# Furthest North and South SNOTELs
get_min_and_max_rows(metadata, "latitude")
```

```python
# Furthest West and East SNOTELs
get_min_and_max_rows(metadata, "longitude")
```

```python
# Oldest/Newest Stations
get_min_and_max_rows(metadata, "begin_date")
```
